### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('chipotle.csv') 
```

```python
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].str.strip('$')
chipo['item_price']
    
```

```python
chipo['item_price'] = chipo['item_price'].astype(float)
chipo['item_price']
```

```python
above_ten_dollar = chipo[chipo['item_price'] > 10]
```

```python
above_ten_dollar.nunique()
```

```python
new_data = chipo[chipo['quantity']==1]
```

```python
new_data['quantity'].shape
```

```python
new_data['item_name'][new_data['item_price']>10].nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
new_data[['item_name','item_price']].drop_duplicates()
```

### Step 6. Sort by the name of the item

```python
new_data.sort_values('item_name',ascending=True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
highest_price = chipo['item_price'].max()
```

```python
highest_price
```

```python
chipo[chipo['item_price']==highest_price]
```
